```python
import pandas as pd
from pathlib import Path
```

```python
df = pd.read_csv(r"refined_data\all_stock_data.csv")
df.head(3)
```

```python
from pathlib import Path

output_dir = Path(r"C:\Users\hp\Desktop\GUVI Projects\Stock Analysis\Ticker_wise_data")
output_dir.mkdir(parents=True, exist_ok=True)

unique_tickers = df['ticker'].unique()

for i in unique_tickers:
    ticker_df = df[df['ticker'] == i]
    
    output_csv = output_dir / f"{i}.csv"
    ticker_df.to_csv(output_csv, index=False)
    
    print(f"CSV created: {output_csv}")
    print(f"{i} : {ticker_df.shape}")
```
