```python
from pathlib import Path
import sys
import yaml
import pandas as pd
from pathlib import Path
```

```python


# Choose one month folder manually for testing
test_month = Path(r"C:\Users\hp\Desktop\GUVI Projects\Stock Analysis\raw_data\2023-10")

# List yaml files inside this month folder
yaml_files = sorted(test_month.glob("*.yaml"))
print(f"Found {len(yaml_files)} YAML files in {test_month.name}")

# Pick the first YAML file for testing
test_file = yaml_files[0]
print(f"Using test file: {test_file.name}")

# Load YAML
with open(test_file, "r") as f:
    yaml_data = yaml.safe_load(f)

print("\nYAML loaded successfully!")
print("Data type:", type(yaml_data))

# Convert YAML to DataFrame
df = pd.DataFrame([yaml_data]) if isinstance(yaml_data, dict) else pd.DataFrame(yaml_data)

df.head()
```

```python

# Test month (same as before)
month_folder = Path(r"C:\Users\hp\Desktop\GUVI Projects\Stock Analysis\raw_data\2023-10")

all_yaml_files = sorted(month_folder.glob("*.yaml"))
print(f"Total YAML files in {month_folder.name}: {len(all_yaml_files)}")

combined_df = []  # temporary list to store individual dfs

for file in all_yaml_files:
    with open(file, "r") as f:
        data = yaml.safe_load(f)

    # Convert dict or list to DataFrame
    df = pd.DataFrame([data]) if isinstance(data, dict) else pd.DataFrame(data)

    combined_df.append(df)

# Combine all rows into one final DataFrame
final_df = pd.concat(combined_df, ignore_index=True)

print("Combined DataFrame shape:", final_df.shape)
final_df.head()
```

```python
# Step 4: Save the combined DataFrame (for one folder) as CSV

output_path = Path(r"C:\Users\hp\Desktop\GUVI Projects\Stock Analysis\combined_2023-10.csv")

final_df.to_csv(output_path, index=False)

print(f"CSV saved successfully at:\n{output_path}")
```

```python

base_dir = Path(r"C:\Users\hp\Desktop\GUVI Projects\Stock Analysis\raw_data")

# Get all subfolders (2023-10, 2023-11, ... 2024-11)
month_folders = sorted([f for f in base_dir.iterdir() if f.is_dir()])

print(f"Total month folders found: {len(month_folders)}")

master_list = []  # will hold all dataframes

for month in month_folders:
    print(f"\nProcessing folder: {month.name}")

    yaml_files = sorted(month.glob("*.yaml"))
    print(f"  Found {len(yaml_files)} YAML files")

    for file in yaml_files:
        with open(file, "r") as f:
            data = yaml.safe_load(f)

        # Convert YAML data to DataFrame
        df = pd.DataFrame([data]) if isinstance(data, dict) else pd.DataFrame(data)

        master_list.append(df)

# Combine all data
master_df = pd.concat(master_list, ignore_index=True)

print("\nFINAL SHAPE OF MASTER DATAFRAME:", master_df.shape)
master_df.head()
```

```python
# Step 6: Save final combined CSV

output_csv = Path(r"C:\Users\hp\Desktop\GUVI Projects\Stock Analysis\all_stock_data.csv")
master_df.to_csv(output_csv, index=False)

print(f"Master CSV created at:\n{output_csv}")
```
