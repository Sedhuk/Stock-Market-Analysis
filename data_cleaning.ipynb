```python
import matplotlib.pyplot as plt

sample = "TCS"  

temp = df[df['Ticker'] == sample]

plt.figure(figsize=(12,5))
plt.plot(temp['date'], temp['close'])
plt.title(f"Closing Price Trend  {sample}")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.grid(True)
plt.show()
```

```python
sample = "SBIN"
temp = df[df['Ticker'] == sample]

plt.figure(figsize=(12,5))
plt.plot(temp['date'], temp['volume'])
plt.title(f"Volume Trend  {sample}")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from pathlib import Path


df = pd.read_csv(r"C:\Users\hp\Desktop\GUVI Projects\Stock Analysis\refined_data\all_stock_data.csv")

# Convert date column to datetime
df['date'] = pd.to_datetime(df['date'])

# Convert month column if needed
df['month'] = pd.to_datetime(df['month']).dt.to_period('M').astype(str)

df.head()

# Step 6: Save final combined CSV

output_csv = Path(r"C:\Users\hp\Desktop\GUVI Projects\Stock Analysis\refined_data\rf_all_stock_data.csv")
df.to_csv(output_csv, index=False)

print(f"Master CSV created at:\n{output_csv}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

close_df = df.pivot(index='date', columns='Ticker', values='close')
corr_matrix = close_df.corr()

plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, cmap="coolwarm", center=0)
plt.title("Stock Closing Price Correlation Heatmap")
plt.show()
```
